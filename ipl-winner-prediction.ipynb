## ðŸ“Š `get_df_info(df, datatypes=[], sort_columns=None, sort_order=None)`

This utility function provides a quick summary of a DataFrame's columns, including data type, number of unique values, and missing value statistics. It also supports filtering by data type and custom sorting.

### ðŸ”§ Parameters:
- **`df`** *(pd.DataFrame)*: The input DataFrame to be profiled.
- **`datatypes`** *(list, optional)*: A list of data types to include (e.g., `['int64', 'float64']`). If empty, all columns are included.
- **`sort_columns`** *(list, optional)*: List of columns in the summary to sort by (e.g., `['Unique Values', 'Null Count']`).
- **`sort_order`** *(list, optional)*: Sorting order for each `sort_columns` item. Accepts `'asc'` or `'desc'` for each (e.g., `['desc', 'asc']`).

### ðŸ“‹ Output:
A DataFrame with the following columns:
- **`Column`**: Column name in the original DataFrame.
- **`Data Type`**: The inferred pandas data type.
- **`Unique Values`**: Number of unique non-null values in the column.
- **`Non-Null Count`**: Number of non-null (non-NaN) entries.
- **`Null Count`**: Number of missing (NaN) entries.

### ðŸ“Œ Example Usage:
```python
# Profile all columns and sort by Null Count descending
get_df_info(df, sort_columns=['Null Count'], sort_order=['desc'])

# Only include numeric columns and sort by Unique Values
get_df_info(df, datatypes=['int64', 'float64'], sort_columns=['Unique Values'], sort_order=['desc'])


````python
def get_df_info(df, datatypes=[], sort_columns=None, sort_order=None):
    lst_tmp = []

    for col in df.columns:
        col_dtype = str(df.dtypes[col])
        if datatypes:
            if col_dtype in datatypes:
                lst_tmp.append([col, col_dtype, df[col].nunique(), df[col].notnull().sum(), df.shape[0] - df[col].notnull().sum()])
        else:
            lst_tmp.append([col, col_dtype, df[col].nunique(), df[col].notnull().sum(), df.shape[0] - df[col].notnull().sum()])

    df_summary = pd.DataFrame(
        lst_tmp,
        columns=["Column", "Data Type", "Unique Values", "Non-Null Count", "Null Count"]
    )

    # Apply sorting if specified
    if sort_columns:
        # Validate sort_columns
        invalid_cols = [col for col in sort_columns if col not in df_summary.columns]
        if invalid_cols:
            raise ValueError(f"Invalid sort column(s): {invalid_cols}")

        # Default to ascending if no sort_order is provided
        if sort_order is None:
            sort_order = ["asc"] * len(sort_columns)

        # Validate lengths match
        if len(sort_order) != len(sort_columns):
            raise ValueError("Length of sort_order must match length of sort_columns")

        # Convert 'asc'/'desc' to True/False
        sort_order_bool = []
        for order in sort_order:
            if isinstance(order, str):
                order = order.lower()
                if order == "asc":
                    sort_order_bool.append(True)
                elif order == "desc":
                    sort_order_bool.append(False)
                else:
                    raise ValueError(f"Invalid sort order: '{order}'. Use 'asc' or 'desc'.")
            else:
                raise ValueError(f"Sort order must be a string ('asc' or 'desc'), got {order}")

        # Perform sorting
        df_summary = df_summary.sort_values(by=sort_columns, ascending=sort_order_bool)

    return df_summary
````
